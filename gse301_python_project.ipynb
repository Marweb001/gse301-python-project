```python
def grade_student(score):
    if score >= 70:
        grade = "A"
    elif score >= 60:
        grade = "B"
    elif score >= 50:
        grade = "C"
    elif score >= 45:
        grade = "D"
    elif score >= 40:
        grade = "E"
    else:
        grade = "F"

    match grade:
        case "A": feedback = "Excellent performance!"
        case "B": feedback = "Very good work."
        case "C": feedback = "Good effort."
        case "D": feedback = "Needs improvement."
        case "E": feedback = "Barely passed."
        case "F": feedback = "Failed. Try harder next time."

    return grade, feedback
```
